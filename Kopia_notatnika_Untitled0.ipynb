```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("zero-shot-classification", model="cross-encoder/nli-deberta-v3-large")
```

```python
classifier = pipeline(
    "zero-shot-classification",
    model="cross-encoder/nli-deberta-v3-large"
)

# Przykładowy tekst
text = "The government is discussing new environmental regulations."

# Kategorie (etykiety)
labels = ["politics", "culture", "sport"]

# Klasyfikacja tekstu do jednej z kategorii
result = classifier(text, candidate_labels=labels, multi_label=False)

print(result)
```

```python
# 7d — test na kilku zdaniach
examples = [
    "The prime minister announced a new economic plan.",
    "The film festival attracted thousands of visitors.",
    "Lewandowski scored a hat-trick in the Champions League.",
    "Parliament voted to amend the constitution."
]

labels_en = ["politics", "culture", "sport"]

for text in examples:
    res = classifier(text, candidate_labels=labels_en, multi_label=False)
    print(f"\n {text}")
    for lbl, sc in zip(res["labels"], res["scores"]):
        print(f"  {lbl:8s} -> {sc:.4f}")
```

```python
pl2en = {"polityka": "politics", "kultura": "culture", "sport": "sport"}

text_pl = "Lewandowski strzelił hat-tricka w Lidze Mistrzów."
labels_pl = ["polityka", "kultura", "sport"]
labels_en = [pl2en[l] for l in labels_pl]

res_en = classifier(text_pl, candidate_labels=labels_en, multi_label=False)

labels_back = [labels_pl[labels_en.index(l)] for l in res_en["labels"]]
res_en["labels"] = labels_back
print(res_en)
```

```python
from transformers import AutoTokenizer
# Load tokenizer (matches the model)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3.5-mini-instruct")
text = "Recommend sci-fi books for me"
# Tokenize
tokens = tokenizer.encode(text)
print(f"Token IDs: {tokens}")
print(f"Number of tokens: {len(tokens)}")
# Decode back to text
decoded = tokenizer.decode(tokens)
print(f"Decoded: {decoded}")
# See individual tokens
token_strings = [tokenizer.decode([t]) for t in tokens]
print(f"Token strings: {token_strings}")
```
